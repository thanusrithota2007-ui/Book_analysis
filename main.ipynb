```python
import pandas as pd 
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv(r"data.csv",encoding="latin1")
df2=pd.read_csv(r"data.csv",encoding="latin1")
df3=pd.read_csv(r"data.csv",encoding="latin1")
```

```python
print(df1.columns)
print(df2.columns)
print(df3.columns)
```

```python
df1["Year"] = 2023
df2["Year"] = 2024
df3["Year"] = 2025

df = pd.concat([df1, df2, df3], ignore_index=True)
```

```python
df.shape
```

```python
df.info()
```

- It is observered that columns like Rating,Price,Reading,Publishing date were in object data type for better analysis they were converted as numeric

```python
df["Rating"] = df["Rating"].str.extract(r"(\d+\.\d+)").astype(float)
```

```python
df["Publishing date"] = pd.to_datetime(df["Publishing date"], errors="coerce")
```

```python
df["price"] = df["price"].str.replace("$", "", regex=False).astype(float)
```

```python
df["Published Year"] = df["Publishing date"].dt.year
```

```python
df.info()
```

```python
df["Rating"].median()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["Rating"], bins=10, kde=True)
plt.title("Distribution of Ratings")
plt.show()
```

- The distribution of ratings is right-skewed towards higher values, with most books having ratings above 4.0.

- This indicates that bestseller books generally maintain high quality or strong audience satisfaction. Extremely low ratings are rare in the dataset.

```python
plt.figure(figsize=(6,4))
sns.histplot(df["price"], bins=15, kde=True)
plt.title("Distribution of Price")
plt.show()
```

-The majority of books are priced below $20.

-There are a few outliers:

-Some books priced above $40

-Very few priced extremely low

-This suggests that bestsellers tend to stay within an affordable range, possibly to maximize sales volume.

```python
plt.figure(figsize=(6,4))
sns.histplot(df["reviews count"], bins=20, kde=True)
plt.title("Distribution of Reviews Count")
plt.show()
```

- The review count distribution is highly left-skewed.

- Most books have moderate reviews, while a few books have extremely high review counts. After applying a log transformation, the distribution appears more normalized.

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=df["reviews count"], y=df["Rating"])
plt.xscale("log")
plt.title("Rating vs Reviews Count (Log Scale)")
plt.show()
```

```python
df[["reviews count", "Rating"]].corr()
```

- Correlation: -0.295

- There is a moderate negative correlation between rating and review count.

- This suggests that books with more reviews tend to have slightly lower ratings.

- Possible explanation:

- Popular books reach a wider audience.

- Larger audience = more diverse opinions.

- More diverse opinions = slightly lower average ratings.

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="price", y="Rating", data=df)
plt.title("Price vs Rating")
plt.show()
```

```python
df[["price", "Rating"]].corr()
```

- Correlation: -0.000611

- There is almost no correlation between price and rating.

- This indicates that higher price does not imply better quality (based on ratings).

```python
genre_rating = df.groupby("Genre")["Rating"].mean().sort_values(ascending=False)

plt.figure(figsize=(8,5))
sns.barplot(x=genre_rating.values, y=genre_rating.index)
plt.title("Average Rating by Genre")
plt.show()
```

- An interesting insight is that Self-help books have a higher average rating compared to Fiction and Romance.

- Possible reasons:

- Self-help books are goal-oriented.

- They attract a more targeted audience.

- Satisfaction may be linked to perceived value rather than entertainment.

```python
year_price = df.groupby("Year")["price"].mean()

plt.figure(figsize=(6,4))
sns.barplot(x=year_price.index, y=year_price.values)
plt.title("Average Price by Year")
plt.show()
```

- There was a slight decrease in average price in 2024, but prices returned to similar levels in 2025.

- This suggests minor pricing fluctuation rather than a consistent trend.
